```python
import pandas as pd
import numpy as np
from PIL import Image
import shutil
import os
```

```python
from ultralytics import YOLO
yolo_model = YOLO(
    r"../classification/best.pt")
```

```python
df = pd.read_csv('../data/subm_test.csv', sep=';')
df.head()
```

```python
ans = []
for obj_id, img_name in df[['object_id', 'img_name']].values:
    img = Image.open(os.path.join("../data/test", str(obj_id), img_name))
    res = yolo_model.predict(img)
    a = [r.probs.data for r in res]
    b = [c.cpu().numpy().tolist() for c in a][0]
    i = b.index(max(b))
    k = res[0].names[i]
    ans += [[obj_id, k, img_name]]
```

```python
dfk = pd.DataFrame(ans, columns=["object_id", "group", "img_name"])
dfk.to_csv("subm_test.csv", sep=';')
```
